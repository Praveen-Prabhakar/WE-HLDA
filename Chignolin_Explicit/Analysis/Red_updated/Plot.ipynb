```python
import numpy
import matplotlib
import matplotlib.pyplot as plt
import sys

# Define our x values
xs = numpy.arange(0,1350,1)

# Load in our rate constants
pre_ys_block  = numpy.loadtxt("./data_files/block_rates.dat", usecols=1)
ys_window     = numpy.loadtxt("./data_files/window_rates.dat", usecols=1)
ys_rolling    = numpy.loadtxt("./data_files/rolling_rates.dat", usecols=1)
ys_red    = numpy.loadtxt("./data_files/red_rates.dat", usecols=1)
# Create figure/axis and plot the three curves
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

#ax.plot(xs, ys_window, linewidth=2, label="window")
#ax.plot(xs, ys_rolling, linewidth=2, label="rolling")
ax.plot(xs, ys_red, linewidth=2, label="RED")
plt.yscale('log')
ax.tick_params(axis='both', labelsize=20)
plt.xlabel("Iterations", fontsize=28)
plt.ylabel("rate constant estimate (s$^{-1}$)", fontsize=28)
#plt.xlim(0, xs[-1])
#plt.ylim(1E3, 1E9)
plt.tight_layout()
plt.legend(fontsize=20)

# Save plot
plt.savefig("./images/rate_constants.pdf")
```
